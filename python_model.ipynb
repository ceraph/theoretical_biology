```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from numpy import ma
```

```python
def prey_deriv(x, y, lamb, alpha):
    b = x
    r = y
    return lamb * b - k * b**2 - alpha * b * r

def pred_deriv(x, y, mu, beta):
    b = x
    r = y
    return -mu * r + beta * b * r
```

```python
lamb  = 3
mu    = 2
alpha = 1
beta  = 0.2 
k     = 0.4
```

```python
max   = 10
steps = .1

Y, X = np.meshgrid(np.arange(0, max, steps), np.arange(0, max, steps))
U, V = np.meshgrid(np.arange(0, max, steps), np.arange(0, max, steps))

for y in range(0, len(Y)):
    for x in range(0, len(X)):
        U[x][y] = prey_deriv(X[x][y], Y[x][y], lamb, alpha)
        V[x][y] = pred_deriv(X[x][y], Y[x][y], mu, beta)
```
